# Project 4-3: Plot the exam data with Seaborn

```python
import pandas as pd
import seaborn as sns
```

## Tasks

```python
# 1
data = pd.read_csv('exams.csv')
data.head()
```

```python
# 2
sns.catplot(data=data, x='parental level of education', kind='count')
```

```python
# 3
g = sns.catplot(data=data, x='parental level of education', kind='count')
for ax in g.axes.flat:
    ax.tick_params('x', labelrotation=30)
```

```python
# 4
sns.scatterplot(data=data, x='reading score', y='writing score', hue='test preparation course')
```

```python
# 5
sns.relplot(data=data, x='reading score', y='writing score', hue='test preparation course')
```

```python
# 6
sns.relplot(data=data, x='reading score', y='writing score', hue='test preparation course', height=3.5, aspect=1.5)
```

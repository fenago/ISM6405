# Project 4-2: Plot the avocado data with Pandas

```python
# https://www.kaggle.com/neuromusic/avocado-prices
import pandas as pd
import seaborn as sns
```

## Tasks

```python
# 1
data = pd.read_csv('avocado.csv')
data.head(5)
```

```python
# 2
sns.scatterplot(data=data, x='AveragePrice', y='Total Volume')
```

```python
# 3
sns.relplot(data=data, kind='scatter', x='AveragePrice', y='Total Volume')
```

```python
# 4
sns.relplot(data=data, kind='scatter', x='AveragePrice', y='Total Volume', hue='year',
           size='Total Volume', sizes=(10,100))
```

```python
# 5
sns.catplot(data=data.query('region in ("West","TotalUS","WestTexNewMexico")'),
            kind='bar', x='year', y='Total Volume', hue='region')
```

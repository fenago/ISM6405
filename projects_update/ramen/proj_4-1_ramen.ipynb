# Project 4-1: Plot the ramen data with Pandas

```python
# https://www.kaggle.com/residentmario/ramen-ratings
import pandas as pd
import seaborn as sns
```

## Tasks

```python
# 1
data = pd.read_csv('ramen-ratings.csv')
data.head()
```

```python
# 2
sns.barplot(data=data, x='Style', y='Stars')
```

```python
# 3
sns.catplot(data=data, x='Style', y='Stars', kind='bar')
```

```python
# 4
g = sns.catplot(data=data, x='Style', y='Stars', kind='bar')
for ax in g.axes.flat:
    ax.set(title='Rating by Ramen Type', xlabel='Ramen Type')
```

```python
# 5
g = sns.catplot(data=data.query('Country in ["Japan","India","Taiwan","USA"]'), 
                x='Style', y='Stars', kind='bar', col='Country', col_wrap=2)
```

```python
# 6
g = sns.catplot(data=data.query('Country in ["Japan","India","Taiwan","USA"]'), 
                x='Style', y='Stars', kind='bar', col='Country', col_wrap=2)
countries = data.query('Country in ["Japan","India","Taiwan","USA"]').Country.unique()
for i, ax in enumerate(g.axes.flat):
    ax.set(title=countries[i] + ' Rating by Ramen Type', xlabel='Ramen Type')
```
